```python
import os

path_gagg = "/home/smallsat/Desktop/grips-impress/geant4-impress-grips-piggyback/build/data-out/position_data/gagg"
path_lyso = "/home/smallsat/Desktop/grips-impress/geant4-impress-grips-piggyback/build/data-out/position_data/lyso"

gaggdir = os.listdir(path_gagg)
lysodir = os.listdir(path_lyso)

print(gaggdir)
print(lysodir)
```

```python
def plot_gagg_hist():
    for fold in gaggdir:
        cut_name = fold.split('-')[0]
        soh.main(path_gagg + '/' + fold, cut_name + '-gagg-positional-hist', cut_name + ' GAGG SiPM Hit Positions')
    return 0

def plot_lyso_hist():
    for fold in lysodir:
        cut_name = fold.split('-')[0]
        soh.main(path_lyso + '/' + fold, cut_name + '-lyso-positional-hist', cut_name + ' LYSO SiPM Hit Positions')
    return 0
```
